```python
def param (set1, b1, b2):
    #Rho 
    rhoP, Sp = pearsonr(set1[b1], set1[b2])
    
    #Rho 
    rho, Sp = spearmanr(set1[b1], set1[b2])
    #Tau
    tau, Kp = kendalltau(set1[b1], set1[b2])

    par1 = [rho, tau, rhoP]
    return par1
```
