```python
def plot1(gdfAM1, title, AMset, n, var, cmap, vmin, vmax): 
    if AMset == 1:
        subtitle = 'Set 1 | S1max-S2conc'
    elif AMset == 2:
        subtitle = 'Set 2 | S2max-S1conc'
    elif AMset == 3:
        subtitle = 'Set 3 | Cmax (C=S1+S2)'
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    fig.suptitle(f'{title} | N={n} years', fontsize=20)

    ax.set_title(subtitle, fontsize=16)
#     regions.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), zorder=0)
    mregions.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), alpha=0.6, zorder=0, label='Regions') #NEW
#     sbasin.plot(ax=ax, color='white', edgecolor='black', linewidth=0.5, alpha=0.6, zorder=1)
    mriver.plot(ax=ax, color='blue', linewidth=1.5, alpha=1, zorder=1, label='Rhine River')
    rivers.plot(ax=ax, color='blue', linewidth=0.4, alpha=0.3, zorder=1, label='Secondary Rivers')
    trib.plot(ax=ax, color='blue', linewidth=0.8, alpha=0.8, zorder=1, label='Tributary Rivers') #NEW
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)
    gdfAM1.plot(gdfAM1[var], ax=ax, marker='o', markersize=100, edgecolors='k', cmap=cmap, zorder=2, vmin=vmin, vmax=vmax, 
                legend=True, legend_kwds={'label': f"{var}", 'orientation': "vertical"}, cax=cax)
    
    texto = []
    for c, ratio in zip(gdfAM1.Confluence, gdfAM1[var]):
        texto.append(f"{c} - rho={ratio:.4f}")
    props = dict(boxstyle='square', facecolor='white')
#     texto = f"{var} min = {rmin:.4f} \n{var} max = {rmax:.4f}"
    ax.text(580000, 5960000, f"{texto[0]} \n{texto[1]} \n{texto[2]} \n{texto[3]} \n{texto[4]} \n{texto[5]} \n{texto[6]}", 
            fontsize=14, ha='center', verticalalignment='top', bbox=props)
    
#     rmin = gdfAM1[var].min()
#     rmax = gdfAM1[var].max()
#     props = dict(boxstyle='square', facecolor='white')
#     ax.text(580000, 5960000, f"{var} min = {rmin:.4f} \n{var} max = {rmax:.4f}", fontsize=14, ha='center', verticalalignment='top', bbox=props)
    
    plt.tight_layout()
    plt.show()
    return

def plot1LOG(gdfAM1, title, AMset, n, var, cmap, vmin, vmax): 
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    fig.suptitle(f'{title} ', fontsize=20) #| N={n} years

#     ax.set_title(subtitle, fontsize=16)
#     regions.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), zorder=0)
    rhine.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), linewidth=0.5, alpha=0.5, zorder=0, label='Regions') #NEW
#     mregions.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), linewidth=0.5, alpha=0.5, zorder=0, label='Regions') #NEW
#     sbasin.plot(ax=ax, color='white', edgecolor='black', linewidth=0.5, alpha=0.5, zorder=1)
    C132.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C106.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C68.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C56.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C49.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C14.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1) 
    C4.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    
    mriver.plot(ax=ax, color='blue', linewidth=1, alpha=1, zorder=1, label='Rhine River')
#     rivers.plot(ax=ax, color='blue', linewidth=0.4, alpha=0.4, zorder=1, label='Secondary Rivers')
    trib.plot(ax=ax, color='blue', linewidth=0.8, alpha=0.8, zorder=1, label='Tributary Rivers') #NEW
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)
    gdfAM1.plot(gdfAM1[var], ax=ax, marker='o', markersize=100, edgecolors='k', cmap=cmap, norm=mcol.LogNorm(vmin=vmin, vmax=vmax), 
                zorder=2, legend=True, legend_kwds={'label': f"{var}", 'orientation': "vertical"}, cax=cax)
#     rmin = gdfAM1[var].min()
#     rmax = gdfAM1[var].max()
    texto = []
    for c, ratio in zip(gdfAM1.Confluence, gdfAM1[var]):
        texto.append(f"{c} - A1/A2={ratio:.4f}")
    props = dict(boxstyle='square', facecolor='white')
#     texto = f"{var} min = {rmin:.4f} \n{var} max = {rmax:.4f}"
    ax.text(580000, 5960000, f"{texto[0]} \n{texto[1]} \n{texto[2]} \n{texto[3]} \n{texto[4]} \n{texto[5]} \n{texto[6]}", 
            fontsize=14, ha='center', verticalalignment='top', bbox=props)
    for x, y, label in zip(gdfAM1.X, gdfAM1.Y, gdfAM1.Confluence):
        ax.annotate(label, xy=(x, y), xytext=(3, 3), fontsize=11, textcoords="offset points", weight='bold', rotation=60, bbox=props, zorder=1) #, bbox=props
    plt.tight_layout()
    plt.show()
    return
```

```python
def plot1conf(gdfAM1, c, title, AMset, n, HYD_rho): 
    if AMset == 1:
        subtitle = 'Set 1 | S1max-S2conc'
    elif AMset == 2:
        subtitle = 'Set 2 | S2max-S1conc'
    elif AMset == 3:
        subtitle = 'Set 3 | Cmax (C=S1+S2)'
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    fig.suptitle(f'{title} ', fontsize=20) #| N={n} years

    ax.set_title(subtitle, fontsize=16)
    
#     regions.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), zorder=0)
    rhine.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), linewidth=0.5, alpha=0.5, zorder=0, label='Regions') #NEW
#     mregions.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), alpha=0.6, zorder=0, label='Regions') #NEW
#     sbasin.plot(ax=ax, color='white', edgecolor='black', linewidth=0.5, alpha=0.6, zorder=1)
    globals()[f'C{c+1}'].plot(ax=ax, color='none', edgecolor='green', linewidth=0.6, alpha=1, zorder=1)
    mriver.plot(ax=ax, color='blue', linewidth=1.5, alpha=1, zorder=1, label='Rhine River')
    rivers.plot(ax=ax, color='blue', linewidth=0.4, alpha=0.2, zorder=1, label='Secondary Rivers')
    trib.plot(ax=ax, color='blue', linewidth=0.8, alpha=0.8, zorder=1, label='Tributary Rivers') #NEW
    divider = make_axes_locatable(ax)
#     cax = divider.append_axes("right", size="5%", pad=0.1)
    vmin, vmax = 0, 1
    cmap = 'Reds'
    gdfAM1.plot(gdfAM1['rho_mean'], ax=ax, marker='o', markersize=100, edgecolors='k', cmap=cmap, zorder=2, vmin=vmin, vmax=vmax)
    vmin, vmax = 0.01, 100
    cmap = 'PiYG'
#     gdfAM1.plot(gdfAM1['A1/A2'], ax=ax, marker='o', markersize=70, edgecolors='k', cmap=cmap, norm=mcol.LogNorm(vmin=vmin, vmax=vmax), 
#                 zorder=2, legend=True, legend_kwds={'label': f"A1/A2", 'orientation': "vertical"}, cax=cax)
#     gdfAM1.plot(gdfAM1['A1/A2'], ax=ax, marker='o', markersize=100, edgecolors='k', cmap=cmap, norm=mcol.LogNorm(vmin=vmin, vmax=vmax), 
#                 zorder=2)
    
    rho = gdfAM1.loc[c, 'rho_mean']
    ar = gdfAM1.loc[c, 'A1/A2']
    props = dict(boxstyle='square', facecolor='white')
    ax.text(580000, 5960000, f"MET_rho = {rho:.4f} \nHYD_rho = {HYD_rho[c]:.4f} \n A1/A2 = {ar:.4f}", fontsize=14, ha='center', verticalalignment='top', bbox=props)
    
    plt.tight_layout()
    plt.show()
    return
```

```python
def plot1LOG(gdfAM1, title, AMset, n, var, cmap, vmin, vmax): 
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    fig.suptitle(f'{title} ', fontsize=20) #| N={n} years

#     ax.set_title(subtitle, fontsize=16)
#     regions.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), zorder=0)
    rhine.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), linewidth=0.5, alpha=0.5, zorder=0, label='Regions') #NEW
#     mregions.plot(ax=ax, color='white', edgecolor='black', figsize=(20, 10), linewidth=0.5, alpha=0.5, zorder=0, label='Regions') #NEW
#     sbasin.plot(ax=ax, color='white', edgecolor='black', linewidth=0.5, alpha=0.5, zorder=1)
    C132.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C106.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C68.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C56.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C49.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    C14.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1) 
    C4.plot(ax=ax, color='none', edgecolor='black', linewidth=0.6, alpha=1, zorder=1)
    
    mriver.plot(ax=ax, color='blue', linewidth=1, alpha=1, zorder=1, label='Rhine River')
#     rivers.plot(ax=ax, color='blue', linewidth=0.4, alpha=0.4, zorder=1, label='Secondary Rivers')
    trib.plot(ax=ax, color='blue', linewidth=0.8, alpha=0.8, zorder=1, label='Tributary Rivers') #NEW
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)
    gdfAM1.plot(gdfAM1[var], ax=ax, marker='o', markersize=70, edgecolors='k', cmap=cmap, norm=mcol.LogNorm(vmin=vmin, vmax=vmax), 
                zorder=2, legend=True, legend_kwds={'label': f"{var}", 'orientation': "vertical"}, cax=cax)
#     rmin = gdfAM1[var].min()
#     rmax = gdfAM1[var].max()
    texto = []
    for c, ratio in zip(gdfAM1.Confluence, gdfAM1[var]):
        texto.append(f"{c} - A1/A2={ratio:.4f}")
    props = dict(boxstyle='square', facecolor='white')
#     texto = f"{var} min = {rmin:.4f} \n{var} max = {rmax:.4f}"
    ax.text(470000, 5960000, f"{texto[0]} \n{texto[1]} \n{texto[2]} \n{texto[3]} \n{texto[4]} \n{texto[5]} \n{texto[6]}", 
            fontsize=14, verticalalignment='top', bbox=props)
    for x, y, label in zip(gdfAM1.X, gdfAM1.Y, gdfAM1.Confluence):
        ax.annotate(label, xy=(x, y), xytext=(3, 3), fontsize=11, textcoords="offset points", weight='bold', rotation=60, bbox=props, zorder=1) #, bbox=props
    plt.tight_layout()
    plt.show()
    return
```

```python
def plot3(gdfAM1, gdfAM2, gdfAM3, title, n, var, cmap, vmin, vmax):
#     vmin = np.min([gdfAM1[var], gdfAM2[var], gdfAM3[var]])
#     vmax = np.max([gdfAM1[var], gdfAM2[var], gdfAM3[var]])
    
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
    fig.suptitle(f'{title} | N={n} years', fontsize=25)

    ax1.set_title('Set 1 | S1max-S2conc', fontsize=20)
#     regions.plot(ax=ax1, color='white', edgecolor='black', figsize=(20, 10), zorder=0)
    mregions.plot(ax=ax1, color='white', edgecolor='black', figsize=(20, 10), zorder=0) #NEW
#     sbasin.plot(ax=ax1, color='white', edgecolor='black', linewidth=0.5, alpha=0.6, zorder=1)
    mriver.plot(ax=ax1, color='blue', linewidth=1, alpha=1, zorder=1)
    rivers.plot(ax=ax1, color='blue', linewidth=0.4, alpha=0.4, zorder=1)
    trib.plot(ax=ax1, color='blue', linewidth=0.8, alpha=0.8, zorder=1) #NEW
    divider = make_axes_locatable(ax1)
    cax = divider.append_axes("right", size="5%", pad=0.1)
    gdfAM1.plot(gdfAM1[var], ax=ax1, marker='o', markersize=110, edgecolors='k', cmap=cmap, zorder=2, vmin=vmin, vmax=vmax, legend=True, legend_kwds={'label': f"{var}",
                            'orientation': "vertical"}, cax=cax)
    rmin = gdfAM1[var].min()
    rmax = gdfAM1[var].max()
    props = dict(boxstyle='square', facecolor='white')
    ax1.text(470000, 5960000, f"{var} min = {rmin:.4f} \n{var} max = {rmax:.4f}", fontsize=12, verticalalignment='top', bbox=props)
    
    #AM2
    ax2.set_title('Set 2 | S2max-S1conc', fontsize=20)
#     regions.plot(ax=ax2, color='white', edgecolor='black', figsize=(20, 10), zorder=0)
    mregions.plot(ax=ax2, color='white', edgecolor='black', figsize=(20, 10), zorder=0) #NEW
#     sbasin.plot(ax=ax2, color='white', edgecolor='black', linewidth=0.5, alpha=0.6, zorder=1)
    mriver.plot(ax=ax2, color='blue', linewidth=1, alpha=1, zorder=1)
    rivers.plot(ax=ax2, color='blue', linewidth=0.4, alpha=0.4, zorder=1)
    trib.plot(ax=ax2, color='blue', linewidth=0.8, alpha=0.8, zorder=1) #NEW
    divider = make_axes_locatable(ax2)
    cax = divider.append_axes("right", size="5%", pad=0.1)
    gdfAM2.plot(gdfAM2[var], ax=ax2, marker='o', markersize=110, edgecolors='k', cmap=cmap, zorder=2, vmin=vmin, vmax=vmax, legend=True, legend_kwds={'label': f"{var}",
                            'orientation': "vertical"}, cax=cax)
    rmin = gdfAM2[var].min()
    rmax = gdfAM2[var].max()
    props = dict(boxstyle='square', facecolor='white')
    ax2.text(470000, 5960000, f"{var} min = {rmin:.4f} \n{var} max = {rmax:.4f}", fontsize=12, verticalalignment='top', bbox=props)
    
    #AM3
    ax3.set_title('Set 3 | Cmax (C=S1+S2)', fontsize=20)
#     regions.plot(ax=ax3, color='white', edgecolor='black', figsize=(20, 10), zorder=0)
    mregions.plot(ax=ax3, color='white', edgecolor='black', figsize=(20, 10), zorder=0) #NEW
#     sbasin.plot(ax=ax3, color='white', edgecolor='black', linewidth=0.5, alpha=0.6, zorder=1)
    mriver.plot(ax=ax3, color='blue', linewidth=1, alpha=1, zorder=1)
    rivers.plot(ax=ax3, color='blue', linewidth=0.4, alpha=0.4, zorder=1)
    trib.plot(ax=ax3, color='blue', linewidth=0.8, alpha=0.8, zorder=1) #NEW
    divider = make_axes_locatable(ax3)
    cax = divider.append_axes("right", size="5%", pad=0.1)
    gdfAM3.plot(gdfAM3[var], ax=ax3, marker='o', markersize=110, edgecolors='k', cmap=cmap, zorder=2, vmin=vmin, vmax=vmax, legend=True, legend_kwds={'label': f"{var}",
                            'orientation': "vertical"}, cax=cax)
    rmin = gdfAM3[var].min()
    rmax = gdfAM3[var].max()
    props = dict(boxstyle='square', facecolor='white')
    ax3.text(470000, 5960000, f"{var} min = {rmin:.4f} \n{var} max = {rmax:.4f}", fontsize=12, verticalalignment='top', bbox=props)
    
    plt.tight_layout()
    plt.show()
    return
```
