```python
def PlotConf1(df1, AMset, n, title1, b1, color, order):
    if AMset == 1:
        titleSet = 'Set 1: S1max-S2conc'
    elif AMset == 2:
        titleSet = 'Set 2: S2max-S1conc'
    elif AMset == 3:
        titleSet = 'Set 3: Cmax (C=S1+S2)'
    
    
    if b1 == 'rho_mean':
        ytext = 0.1
        Title = 'Correlations per confluence'
    elif b1 == 'rho/area':
        ytext = 0.001
        Title = 'Correlation/Area per confluence'
    elif b1 == 'CV':
        ytext = 0.05
        Title = 'Coefficient of variation per confluence'
    elif b1 == 'rho_std':
        ytext = 0.1
        Title = 'STD of the correlations per confluence'
    elif b1 == 'S1avg_mean':
        ytext = 0.5
        Title = 'Stream1_peaks/seasonal_mean per confluence'
    elif b1 == 'S2avg_mean':
        ytext = 0.5
        Title = 'Stream2_peaks/seasonal_mean per confluence'
    
    if order == 'Regions':
        plt.figure(figsize=(18, 5)) 

        plt.plot(df1['Confluence'], df1[b1], 'go', label=f'{title1}')

        plt.ylim(ymin = 0)#, ymax = 1)

        plt.text('C1', ytext, 'Alpine',fontsize = 16, rotation = 90)
        plt.axvline(x='C3', color='cadetblue')
        plt.text('C13', ytext, 'High',fontsize = 16, rotation = 90)
        plt.axvline(x='C15', color='cadetblue')
        plt.text('C67', ytext, 'Upper',fontsize = 16, rotation = 90)
        plt.axvline(x='C69', color='cadetblue')
        plt.text('C105', ytext, 'Middle',fontsize = 16, rotation = 90)
        plt.axvline(x='C107', color='cadetblue')
        plt.text('C131', ytext, 'Low',fontsize = 16, rotation = 90)
        plt.axvline(x='C133', color='cadetblue')
        # # plt.text('C133', 0.1, 'Delta',fontsize = 16, rotation = 90)

        df = df1
        ticks = df['Confluence']
        labels = [df['Confluence'][0], '', '', '',df['Confluence'][4], '', '', '', '',df['Confluence'][9], '', '', '','', 
                  df['Confluence'][14], '', '', '','', df['Confluence'][19], '', '','', '', df['Confluence'][24], '', '', '', '', 
                  df['Confluence'][29], '', '', '', '',df['Confluence'][34], '', '', '', '',df['Confluence'][39], '', '', '','', 
                  df['Confluence'][44], '', '', '','', df['Confluence'][49], '', '','', '', df['Confluence'][54], '', '', '', '', 
                  df['Confluence'][59], '', '', '', '',df['Confluence'][64], '', '', '', '',df['Confluence'][69], '', '', '','', 
                  df['Confluence'][74], '', '', '','', df['Confluence'][79], '', '','', '', df['Confluence'][84], '', '', '', '', 
                  df['Confluence'][89], '', '', '', '',df['Confluence'][94], '', '', '', '',df['Confluence'][99], '', '', '','', 
                  df['Confluence'][104], '', '', '','', df['Confluence'][109], '', '','', '', df['Confluence'][114], '', '', '', '', 
                  df['Confluence'][119], '', '', '', '', df['Confluence'][124], '', '', '', '',df['Confluence'][129], '', '', df['Confluence'][132]]
        plt.xticks(ticks=ticks, labels=labels, fontsize = 14, rotation = 90)
        plt.yticks(fontsize = 14)
        plt.title(f'{Title} | {titleSet} ', fontsize = 22) #| N = {n} 
        plt.xlabel('Confluence (Regions: upstream to downstream)', fontsize = 16)
        plt.ylabel(b1, fontsize = 16) 
        plt.legend(fontsize=16, bbox_to_anchor=(1,1), loc="upper left")
        plt.grid(axis='y')
        plt.tight_layout()
    
    elif order == 'Area':
        df1 = df1.sort_values(by="Area")

        plt.figure(figsize=(18, 5)) 

        plt.plot(df1['Confluence'], df1[b1], 'go', label=f'{title1}')
        
        plt.ylim(ymin = 0)#, ymax = 1)

        ticks = df1['Confluence']
        lab = df1.sort_values(by="Area")['Confluence'].values
        labels = [lab[0], '', '', '',lab[4], '', '', '', '',lab[9], '', '', '','', 
                  lab[14], '', '', '','', lab[19], '', '','', '', lab[24], '', '', '', '', 
                  lab[29], '', '', '', '',lab[34], '', '', '', '',lab[39], '', '', '','', 
                  lab[44], '', '', '','', lab[49], '', '','', '', lab[54], '', '', '', '', 
                  lab[59], '', '', '', '',lab[64], '', '', '', '',lab[69], '', '', '','', 
                  lab[74], '', '', '','', lab[79], '', '','', '', lab[84], '', '', '', '', 
                  lab[89], '', '', '', '',lab[94], '', '', '', '',lab[99], '', '', '','', 
                  lab[104], '', '', '','', lab[109], '', '','', '', lab[114], '', '', '', '', 
                  lab[119], '', '', '', '', lab[124], '', '', '', '',lab[129], '', '', lab[132]]
        plt.xticks(ticks=ticks, labels=labels, fontsize = 14, rotation = 90)
        plt.yticks(fontsize = 14)
        plt.title(f'{Title} | {titleSet}', fontsize = 22) # | N = {n} 
        plt.xlabel('Confluence (Area: smallest to largest)', fontsize = 16)
        plt.ylabel(b1, fontsize = 16) 
        plt.legend(fontsize=16, bbox_to_anchor=(1,1), loc="upper left")
        plt.grid(axis='y') #
        plt.tight_layout()
    
    elif order == 'MainRiver':
        plt.figure(figsize=(18, 5)) 

        for i in range(len(df1)):
            if df1['MainRiver'][i] == 1:
                plt.plot(df1['Confluence'][i], df1[b1][i], 'go')#, label=f'MET | {titleSet}'

        plt.ylim(ymin = 0)#, ymax = 1)

        plt.text('C2', ytext, 'Alpine',fontsize = 16, rotation = 90)
        plt.axvline(x='C3', color='cadetblue')
        plt.text('C14', ytext, 'High',fontsize = 16, rotation = 90)
        plt.axvline(x='C15', color='cadetblue')
        plt.text('C67', ytext, 'Upper',fontsize = 16, rotation = 90)
        plt.axvline(x='C68', color='cadetblue')
        plt.text('C106', ytext, 'Middle',fontsize = 16, rotation = 90)
        plt.axvline(x='C107', color='cadetblue')
        plt.text('C132', ytext, 'Low',fontsize = 16, rotation = 90)
        plt.axvline(x='C133', color='cadetblue')
        
        plt.xticks(fontsize = 14, rotation = 90)
        plt.yticks(fontsize = 14)

        plt.title(f'{Title} | {titleSet}', fontsize = 22) #| N = {n} 
        plt.xlabel('Confluence (MainRiver: upstream to downstream)', fontsize = 16)
        plt.ylabel(b1, fontsize = 16) 

        legend_elements = [Line2D([0], [0], marker='o', color='w', label=f'{title1}', markerfacecolor='g', markersize=10)]
        plt.legend(handles=legend_elements, fontsize=16, bbox_to_anchor=(1,1), loc="upper left")
        plt.grid(axis='y')
        plt.tight_layout()
        
    return
```
