## Rho vs Area (and Aratio)

```python
def rhoScatter(df1, AMset, n, Title, b1, b2, color):
    if AMset == 1:
        Legend1 = 'Set 1: S1max-S2conc'
    elif AMset == 2:
        Legend1 = 'Set 2: S2max-S1conc'
    elif AMset == 3:
        Legend1 = 'Set 3: Cmax (C=S1+S2)'
    
    
    plt.figure(figsize=(7, 7)) 
    plt.plot(df1[b1], df1[b2], 'o', color=color)
    plt.ylim(ymin=-0.02, ymax = 1.02)
    plt.xticks(fontsize = 12, rotation = 90)
    plt.yticks(fontsize = 12)
    plt.title(f'{Title} \n {Legend1}', fontsize = 20) 
    plt.xlabel(b1, fontsize = 18)
    plt.ylabel(b2, fontsize = 18) 
#     plt.legend(fontsize=18, bbox_to_anchor=(1,1), loc="upper left")
    plt.grid()    
        
    return

def rhoScatter2(df1, AMset, n, Title, b1, b2, cmap):
    if AMset == 1:
        Legend1 = 'Set 1: S1max-S2conc'
    elif AMset == 2:
        Legend1 = 'Set 2: S2max-S1conc'
    elif AMset == 3:
        Legend1 = 'Set 3: Cmax (C=S1+S2)'
    
    
    vmin=1 #0.001 
    vmax=1000
    
    plt.figure(figsize=(9, 7)) 
#     plt.plot(df1[b1], df1[b2], 'o', color=color, label=f'{Legend1}')
    plt.scatter(df1[b1], df1[b2], c=df1['A1/A2'], cmap=cmap, norm=mcol.LogNorm(vmin=vmin, vmax=vmax), s=80, edgecolors='k', zorder=2)
    plt.colorbar(label="Area1/Area2 Ratio", orientation="vertical")
    plt.ylim(ymin=-0.02, ymax = 1.02)
    plt.xticks(fontsize = 12, rotation = 90)
    plt.yticks(fontsize = 12)
    plt.title(f'{Title} \n {Legend1}', fontsize = 20) 
    plt.xlabel(b1, fontsize = 18)
    plt.ylabel(b2, fontsize = 18) 
#     plt.legend(fontsize=18, bbox_to_anchor=(1,1), loc="upper left")
    plt.grid()    
        
    return

def rhoScatterLOG(df1, AMset, n, Title, b1, b2, color):
    if AMset == 1:
        Legend1 = 'Set 1: S1max-S2conc'
    elif AMset == 2:
        Legend1 = 'Set 2: S2max-S1conc'
    elif AMset == 3:
        Legend1 = 'Set 3: Cmax (C=S1+S2)'
    
    plt.figure(figsize=(7, 7)) 
    plt.plot(df1[b1], df1[b2], 'o', color=color)
    plt.ylim(ymin=-0.02, ymax = 1.02)
    plt.xticks(fontsize = 12, rotation = 90)
    plt.yticks(fontsize = 12)
    plt.title(f'{Title} \n {Legend1} ', fontsize = 20) 
    plt.xlabel(b1, fontsize = 18)
    plt.ylabel(b2, fontsize = 18) 
#     plt.legend(fontsize=18, bbox_to_anchor=(1,1), loc="upper left")
    plt.grid()   
    plt.xscale('log')
        
    return

def rhoScatterLOG2(df1, AMset, n, Title, b1, b2, cmap):
    if AMset == 1:
        Legend1 = 'Set 1: S1max-S2conc'
    elif AMset == 2:
        Legend1 = 'Set 2: S2max-S1conc'
    elif AMset == 3:
        Legend1 = 'Set 3: Cmax (C=S1+S2)'
    
    
    vmin=0.001 
    vmax=1000
    
    plt.figure(figsize=(7, 9)) 
    plt.scatter(df1[b1], df1[b2], c=df1[b1], cmap=cmap, norm=mcol.LogNorm(vmin=vmin, vmax=vmax), s=80, edgecolors='k', zorder=2)
    plt.colorbar(label="Area1/Area2 Ratio", orientation="horizontal")
    plt.ylim(ymin=-0.02, ymax = 1.02)
    plt.xticks(fontsize = 12, rotation = 90)
    plt.yticks(fontsize = 12)
    plt.title(f'{Title}  \n {Legend1}', fontsize = 20) 
    plt.xlabel(b1, fontsize = 18)
    plt.ylabel(b2, fontsize = 18) 
#     plt.legend(fontsize=18, bbox_to_anchor=(1,1), loc="upper left")
    plt.grid(zorder=0)   
    plt.xscale('log')
        
    return
```

```python
def plotS3all(domain, extra, n, b1, b2, xmax, ymax, Title, blue=True): #, xmax, ymax
    plt.figure(figsize=(20, 5))
    plt.suptitle(f'{Title} ', fontsize=20, y=1.05) #\n | {b1} vs {b2}
    if blue==True:
        markers = ['o', 'o', 'o', 'o', 'o', 'o', 'o']
        colors = ['b', 'b', 'b', 'b', 'b', 'b', 'b']
    else:
        markers = ['o', 'o', 'o', 'o', 'o', 'o', 'o']
        colors = sns.color_palette("plasma_r", 7)
        # colors = ['skyblue', 'mediumblue', 'y', 'forestgreen', 'lightcoral', 'maroon', 'orange']

    #AM1
    plt.subplot(1,3,1)
    A = 1
    for i in range(len(n)):
        plt.plot(globals()[f'{domain}_N{n[i]}_AM{A}_{extra}'][b1], globals()[f'{domain}_N{n[i]}_AM{A}_{extra}'][b2], markers[i], color=colors[i], label=f'N={n[i]}')
#         plt.plot(globals()[f'gpd_AM1_n{n[i]}{extra}'][b1], globals()[f'gpd_AM1_n{n[i]}{extra}'][b2], markers[i], label=f'N={n[i]}')
    plt.title(f'Set 1 | MSmax-TSconc', fontsize=18) #runoff
    plt.xlabel(f'Spearmans r_s per confluence', fontsize=14)
    plt.xticks(rotation = 0, fontsize=12)
    plt.ylabel(f'Standard deviation', fontsize=14) #Q m3/s
    plt.yticks(fontsize=12)
    plt.xlim(left=0, right=xmax)
    plt.ylim(ymin = 0, ymax = ymax)
    # plt.xscale('log')
    plt.legend()
    plt.grid()

    #AM2
    plt.subplot(1,3,2)
    A = 2
    for i in range(len(n)):
        plt.plot(globals()[f'{domain}_N{n[i]}_AM{A}_{extra}'][b1], globals()[f'{domain}_N{n[i]}_AM{A}_{extra}'][b2], markers[i], color=colors[i], label=f'N={n[i]}')
#         plt.plot(globals()[f'gpd_AM2_n{n[i]}{extra}'][b1], globals()[f'gpd_AM2_n{n[i]}{extra}'][b2], markers[i], label=f'N={n[i]}')
    plt.title(f'Set 2 | TSmax-MSconc', fontsize=18) #runoff
    plt.xlabel(f'Spearmans r_s per confluence', fontsize=14)
    plt.xticks(rotation = 0, fontsize=12)
    plt.ylabel(f'Standard deviation', fontsize=14) #Q m3/s
    plt.yticks(fontsize=12)
    plt.xlim(left=0, right=xmax)
    plt.ylim(ymin = 0, ymax = ymax)
    # plt.xscale('log')
    plt.legend()
    plt.grid()

    #AM3
    plt.subplot(1,3,3)
    A = 3
    for i in range(len(n)):
        plt.plot(globals()[f'{domain}_N{n[i]}_AM{A}_{extra}'][b1], globals()[f'{domain}_N{n[i]}_AM{A}_{extra}'][b2], markers[i], color=colors[i], label=f'N={n[i]}')
#         plt.plot(globals()[f'gpd_AM2_n{n[i]}{extra}'][b1], globals()[f'gpd_AM2_n{n[i]}{extra}'][b2], markers[i], label=f'N={n[i]}')
    plt.title(f'Set 3 | Cmax (C=MS+TS)', fontsize=18) #runoff
    plt.xlabel(f'Spearmans r_s per confluence', fontsize=14)
    plt.xticks(rotation = 0, fontsize=12)
    plt.ylabel(f'Standard deviation', fontsize=14) #Q m3/s
    plt.yticks(fontsize=12)
    plt.xlim(left=0, right=xmax)
    plt.ylim(ymin = 0, ymax = ymax)
    # plt.xscale('log')
    plt.legend()
    plt.grid()
```
